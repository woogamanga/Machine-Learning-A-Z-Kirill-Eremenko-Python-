# Multiple Linear Regression

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('50_Startups.csv')
```

```python
type(dataset)
```

```python
dataset.head()
```

```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
print(X) #this is our matrix of features
```

```python
print(y) #this is our dependent variable vector 
```

## Encoding categorical data

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [3])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```python
print(X) #3 dummy variables have been created to represent New York, California and Florida 
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
print(X_train) #this is 80% of the dataset 
```

```python
print(X_test) #this is 20% of the dataset 
```

```python
print(y_train) #this is 80% of the dataset 
```

```python
print(y_test) #this is 20% of the dataset 
```

## Training the Multiple Linear Regression model on the Training set

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

## Predicting the Test set results

```python
#since we can't visualise the line-of-best-fit as we would need a plane beyond 2D
#the only thing we can do is to concatenate two vectors
#1 vector contains the predicted/calculated outputs
#the other vector contains the actual outputs (otherwise known as the Ground Truth)
#y_pred is our vector of predicted outputs
y_pred = regressor.predict(X_test)
np.set_printoptions(precision=2) #discard the imaginary portion essentially
#reshaping the 2 vectors so that it displays as a column vector instead of a row vector by default
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))

#the array below shows predicted output (left) vs actual output (right)
#future lessons will cover more on performance metrics
```

# K-Nearest Neighbors (K-NN)

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('Social_Network_Ads.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
print(X) #this is our matrix of features
```

```python
print(y) #this is our dependent variable vector 
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0) #fixing the seed here for learning purposes
```

```python
print(X_train) #this is 75% of our feature matrix 
```

```python
print(y_train) #this is 75% of our dependent variable vector
```

```python
print(X_test) #this is 25% of our feature matrix 
```

```python
print(y_test) #this is 25% of our dependent variable vector
```

## Feature Scaling

```python
#as usual we perform scaling (using standardization) after we have split the dataset
#into training and test sets
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test) #note the use of transform instead of fit_transform here 
```

```python
print(X_train)
```

```python
print(X_test)
```

## Training the K-NN model on the Training set (the code template is very similar to logistic regression - only this cell needs to be changed***)

```python
#import the KNeighborsClassifier class from the neighbors module
from sklearn.neighbors import KNeighborsClassifier
#create an object of the KNeighborsClassifier class 
#K = 5 in this case 
#metric is set to minkowski (Generalisation of both Manhattan and Euclidean distances)
#p is the power parameter of the minkowski metric 
#p=2 is equivalent to using Euclidean distance while p=1 is equivalent to using Manhattan distance, under Minkowski
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
#fit the KNeighborsClassifier (K-NN) model on the dataset
classifier.fit(X_train, y_train)
```

## Predicting a new result

```python
print(classifier.predict(sc.transform([[30,87000]]))) #perform scaling before using predict method of the classifier
```

## Predicting the Test set results

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
#we are concatenating two 2D array by the column basis 
#the first column shows the predicted outputs using the trained classifier
#the second column shows the actual outputs 
#we didn't do np.set_printoptions here because we're not doing any rounding/truncation
```

## Making the Confusion Matrix

```python
#similar to the one from logistic regression
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
#format of the confusion matrix as per given in sklearn is such that
#TN = 64
#FP = 4
#FN = 3
#TP = 29
print(cm)
#accuracy = (TN + TP)/ (FN + FP + TN + TP)
#.        = (64 + 29)/ (3 + 4 + 64 + 29) = 93/100 = 0.93
#
accuracy_score(y_test, y_pred)
```

## Visualising the Training set results (note that the prediction line/decision boundary is not a straight line like it was in logistic regression, which makes it a better classifer)

```python
from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(X_train), y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 1),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 1))
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('K-NN (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

## Visualising the Test set results

```python
from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(X_test), y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 1),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 1))
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('K-NN (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
